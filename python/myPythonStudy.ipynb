#### 1.2. 파이썬 기본 문법
* 주석
  * 주석은 코드에 대한 설명이나 메모를 추가하는 방법입니다. 주석은 실행되지 않으며 파이썬 인터프리터에 의해 무시됩니다. <br>
  주석을 추가하려면 해시기호 (#)을 사용합니다.

  ~~~
  # 이것은 주석입니다.
  print("Hello, world!")
  ~~~
  주석은 코드를 읽고 이해하는 데 도움이 됩니다. 또한 코드의 버그를 찾는 데 도움이 될 수 있습니다.

  주석을 사용하는 몇 가지 일반적인 방법은 다음과 같습니다.

  * 코드 블록의 목적을 설명합니다.
  * 변수 또는 함수의 사용법을 설명합니다.
  * 버그를 수정하는 방법을 설명합니다.
  * 코드의 특정 부분을 테스트하는 방법을 설명합니다.
  * 주석은 코드를 더 읽기 쉽고 유지 관리하기 쉽게 만드는 좋은 방법입니다.

  다음은 주석을 사용하는 몇 가지 예입니다.
  ~~~
  # 이 함수는 두 개의 숫자를 더합니다.
  def add(a, b):
    return a + b

  # 이 코드는 사용자가 입력한 두 개의 숫자를 더합니다.
  a = int(input("첫 번째 숫자를 입력하세요: "))
  b = int(input("두 번째 숫자를 입력하세요: "))
  print(add(a, b))
  ~~~

  이 코드는 다음과 같이 출력됩니다.
  ~~~
  첫 번째 숫자를 입력하세요: 5
  두 번째 숫자를 입력하세요: 6
  11
  ~~~

다음은 주석을 사용하는 몇 가지 예입니다.

* 변수, 자료형, 연산자
  * 변수, 자료형, 연산자
    * 변수는 값을 저장하는 공간입니다. 변수를 만들려면   다음과 같이 합니다.
    ~~~
    x = 5
    ~~~
  * 자료형은 변수에 저장된 값의 유형을 나타냅니다. 파이썬에는 다음과 같은 기본 자료형이 있습니다.
    * 정수 (int)
    * 실수 (float)
    * 문자열 (str)
    * 불린 (bool)
  * 연산자는 변수와 값을 조작하는 데 사용됩니다. 파이썬에는 다음과 같은 연산자가 있습니다.
    * 산술 연산자 (+,-,*,/,%)
    * 비교 연산자 (==, !=, <, >, <=, >=)
    * 논리 연산자 (and, or, not)
* 제어 흐름 (조건문, 반복문)
  * 조건문은 코드의 흐름을 제어하는 데 사용됩니다. 가장 일반적인 조건문은 if, elif, else 입니다.
  
  ~~~
  if x > 0:
  print("x는 양수입니다.")
    elif x < 0:
  print("x는 음수입니다.")
    else:
  print("x는 0입니다.")
  ~~~

  * 반복문은 코드 블록을 반복하는 데 사용됩니다. 가장 일반적인 반복문은 for 와 while 입니다.
  
  ~~~
  # 1부터 10까지 출력합니다.
  for i in range(1, 11):
    print(i)

  # x가 0이 될 때까지 반복합니다.
  while x > 0:
    print(x)
    x -= 1
  ~~~
* 함수
  * 함수는 코드 블록에 이름을 지정하는 방법입니다. 함수를 만들려면 다음과 같이 합니다.

  ~~~
  def my_function(x):
    return x + 1

  print(my_function(5))  # 6
  ~~~

* 모듈 및 패키지

  1. 모듈
  * 모듈은 Python 코드를 하나의 파일에 묶어놓은 것입니다. 모듈을 사용하면 코드를 재사용하고 관리하기 쉽습니다.
  
  모듈 사용 방법

  * import 명령어를 사용하여 모듈을 불러옵니다.
  * 모듈의 함수, 변수, 클래스 등을 사용할 수 있습니다.

  ~~~
  import 모듈명

  # 모듈의 함수 사용
  모듈명.함수명()

  # 모듈의 변수 사용
  변수 = 모듈명.변수명
  ~~~

  2. 패키지
  * 패키지는 여러 개의 모듈을 하나의 묶음으로 관리하는 방법입니다. 패키지를 사용하면 코드를 더욱 효율적으로 관리할 수 있습니다.

  패키지 만들기

  * __init__.py 파일을 만들어 패키지를 정의합니다.
  * 패키지 내 모듈을 __init__.py 파일에 등록합니다.
  패키지 사용 방법

  * import 명령어를 사용하여 패키지를 불러옵니다.
  * 패키지 내 모듈을 패키지 이름을 통해 사용할 수 있습니다.

  ~~~
  import 패키지명

  # 패키지 내 모듈 사용
  패키지명.모듈명.함수명()
  ~~~

  3. 모듈 vs. 패키지

| 구분 | 정의 | 용도 | 사용 방법 |
|---|---|---|---|
| 모듈 | 특정 기능을 수행하도록 설계된 코드 단위 | 코드 재사용 및 관리 | `import 모듈명` |
| 패키지 | 모듈, 데이터, 설정 파일 등을 논리적으로 그룹화한 것 | 코드 관리 및 코드의 이름 공간 관리 | `import 패키지명` |
